```python
import pandas as pd
```

```python
titles = pd.read_csv('../data/netflix_titles.csv', index_col='show_id')
titles.head()
```

# Null values

```python
titles[titles.columns[titles.isnull().any()]].isnull().sum()
```

```python
titles[titles['duration'].isnull()].head()
```

```python
idx = titles['duration'].isnull()
titles.loc[idx, 'duration'] = titles['rating']
titles.loc[idx, 'rating'] = 'TV-MA'

titles.loc['s5542']
```

```python
titles[titles['rating'].isnull()].head()
```

```python
titles.loc[titles['rating'].isnull(), 'rating'] = 'NR'
titles.loc["s7538", 'rating'] = 'TV-14'
titles.loc["s7313", 'rating'] = 'TV-MA'
titles.loc["s7538", 'rating'] = 'PG-13'
titles[titles['rating'].isnull()].head()
```

```python
titles[titles.rating == "UR"]
```

```python
titles.loc[titles['rating'] == 'UR', 'rating'] = 'NR'
titles.loc["s8791", "rating"] = "PG-13"
```
